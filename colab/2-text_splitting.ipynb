Text Splitting

```python
# Steps
# 1. Load PDF
# 2. Split PDF (2 types -> CharacterTextSplitter, RecursiveCharacterTextSplitter)
```

```python
# 1. Load PDF
from langchain_community.document_loaders import PyPDFLoader
```

```python
pdfDoc = PyPDFLoader("sample_pdf.pdf")
```

```python
doc = pdfDoc.load()
```

```python
# 2. Split PDF - using CharacterTextSplitter
from langchain_text_splitters import CharacterTextSplitter
```

```python
pip install tiktoken
```

```python
characterTextSplitter = CharacterTextSplitter.from_tiktoken_encoder(chunk_size=50, chunk_overlap=0)
```

```python
doc
```

```python
consolidated_pages = '\n'.join([page.page_content for page in doc])
```

```python
characterTextSplitter.split_text(consolidated_pages)
```

```python
# 2. Split PDF - using RecucrsiveCharacterTextSplitter
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
recursiveCharacterTextSplitter = RecursiveCharacterTextSplitter(chunk_size=50, chunk_overlap=0)
```

```python
recursiveCharacterTextSplitter.split_text(consolidated_pages)
```
