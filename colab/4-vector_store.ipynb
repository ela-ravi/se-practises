```python
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
huggingFaceEmbeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
from langchain_community.vectorstores import FAISS
```

```python
texts = [
    "What is the capital of France?",
    "What is the capital of Germany?",
    "What is the capital of Italy?",
]
```

```python
faiss_store = FAISS.from_texts(texts=texts, embedding=huggingFaceEmbeddings)
```

```python
faiss_store.save_local("faiss_store")
```

```python
retriever = faiss_store.as_retriever()
```

```python
query = 'France'
```

```python
document = retriever.invoke(query)
```

```python
print(document)
```

```python
for doc in document:
    print(doc.page_content)
```

Steps:

1. Initialize document (.txt, .pdf, webpageUrl, arxiv_doc_with_id, wikipedia)
2. Load document
3. Extract text from all pages in the loaded document and consolidate them into a single string
4. Split the consolidated string into chunks (text splitting)
5. Embed all chunks
6. Store in Vector store
7. For user to retrieve info from vectore store, cast vector store as retriever
