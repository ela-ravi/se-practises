```python
# 4. Load contents from Arxiv Research papers
from langchain_community.document_loaders import ArxivLoader # pre-requisite to use ArxivLoader are 'arxiv' and 'PyMuPDF'. Hence install them
```

```python
pip install arxiv PyMuPDF
```

```python
arxivDoc = ArxivLoader(query="2312.10997")
```

```python
arxivDoc.load() # This will download the arxiv paper to your local with file name of what is queried. eg in this case it is "2312.10997v5.Retrieval_Augmented_Generation_for_Large_Language_Models__A_Survey"
```
