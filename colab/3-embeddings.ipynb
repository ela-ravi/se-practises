```python
# 1. OpenAIEmbedding
from langchain_openai import OpenAIEmbeddings
```

```python
openAIEmbeddings = OpenAIEmbeddings('api_key') # provide openai api key
```

```python
text = "Hi, This is Elamurugan. How do you do?"
```

```python
openAIEmbeddings.embed_query(text)
```

```python
# 2. Huggingface Embeddings
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
huggingFaceEmbeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
text = "Hi, This is Elamurugan. How do you do?"
```

```python
huggingFaceEmbeddings.embed_query(text)
```

Steps:

1. Initialize document (.txt, .pdf, webpageUrl, arxiv_doc_with_id, wikipedia)
2. Load document
3. Extract text from all pages in the loaded document and consolidate them into a single string
4. Split the consolidated string into chunks (text splitting)
5. Embed all chunks
